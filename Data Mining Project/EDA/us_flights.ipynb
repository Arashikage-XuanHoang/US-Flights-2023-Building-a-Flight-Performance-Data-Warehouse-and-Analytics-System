```python
df ["FlightDate"] = pd.to_datetime (df ["FlightDate"])
df ["month"] = df ["FlightDate"].dt.month
df ["day"] = df ["FlightDate"].dt.day
# df ["year"] = df ["FlightDate"].dt.year --> tgian phân bổ trong 2023
df ["quarter"] = df ["FlightDate"].dt.quarter
```

```python
def label_delay(x):
    if x < 0:
        return "early"
    elif x == 0:
        return "on_time"
    elif 0 < x < 15:
        return "late"
    else:  # x >= 15
        return "extreme_late"
```

## c. Numerical

### Departure Delay

```python
def plot_dep_delay(data, x, color="#A868D2", xlabel=None, title=None, bins=50):
    """
    Vẽ histogram + đường KDE cho cột x trong dataframe.

    Tham số:
    ----------
    data : DataFrame
        Dữ liệu đầu vào (ví dụ dep_delay_filtered)
    x : str
        Tên cột cần vẽ
    color : str, optional
        Mã màu cho histogram (mặc định: tím nhạt #A868D2)
    xlabel : str, optional
        Nhãn trục X
    title : str, optional
        Tiêu đề biểu đồ
    bins : int, optional
        Số lượng cột trong histogram (mặc định 50)
    """

    plt.figure(figsize=(8, 5))
    sns.histplot(
        x=x,
        data=data,
        bins=bins,
        stat="density",
        color=color,
        edgecolor="white",
        alpha=0.7
    )
    sns.kdeplot(data[x], color="black")

    plt.xlabel(xlabel if xlabel else x)
    plt.title(title if title else f"Histogram of {x}")
    plt.tight_layout()
    plt.show()
```

```python
dep_delay_filtered = df.loc [df ["Dep_Delay"].between (-60, 300)]
plot_dep_delay(
    data=dep_delay_filtered,
    x="Dep_Delay",
    color="#FFB84D",
    xlabel="Departure Delay (minutes)",
    title="Histogram of Departure Delay"
)

sns.boxplot (
    x="Dep_Delay", data=df,
)
```

# 2. Bivariate Analysis

```python
delay_columns = ['Delay_Carrier', 'Delay_Weather', 'Delay_NAS', 'Delay_Security', 'Delay_LastAircraft']
count_factor_delay = pd.DataFrame({
    col: [len(df[df[col] > 15])] for col in delay_columns
}, index=['Count']).transpose()

delay_columns = ['Delay_Carrier', 'Delay_Weather', 'Delay_NAS', 'Delay_Security', 'Delay_LastAircraft']
mean_factor_delay = pd.DataFrame({
    col: [df[col].mean()] for col in delay_columns
}, index=['Mean']).transpose()

fig, ax = plt.subplots (1,2, figsize=(16,6))
ax[0].bar(
    x=count_factor_delay.index,
    height=count_factor_delay.Count,
    color="skyblue"
)
ax[0].set_title("Count of time delay type")
ax[0].set_ylabel('Count')
ax[0].tick_params(axis='x', rotation=45)

ax[1].bar(x=mean_factor_delay.index,height=mean_factor_delay.Mean)
ax[1].set_title("Average delay time by time delay type")
ax[1].set_ylabel('Average delay time (minutes)')
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout ()
plt.show()
```

```python
fig, ax = plt.subplots (1,2, figsize=(12,5))
sns.boxplot(
    x="Dep_Delay_Tag",
    y="Flight_Duration",
    data=df,
    palette="Set2", ax=ax[0]
)
ax[0].set_title("Flight Duration vs. Delay Tag")
ax[0].set_xlabel("Delay Tag (0 = No Delay, 1 = Delay)")
ax[0].set_ylabel("Flight Duration (minutes)")
ax[0].grid ()

sns.violinplot(
    x="Dep_Delay_Tag",
    y="Flight_Duration",
    data=df,
    palette="pastel",
    inner="quartile",
    ax=ax[1]
)
ax[1].set_title("Distribution of Flight Duration by Delay Tag")
ax[1].set_xlabel("Delay Tag (0 = No Delay, 1 = Delay)")
ax[1].set_ylabel("Flight Duration (minutes)")
ax[1].grid ()

plt.tight_layout
plt.show()
```

```python
def plot_delay_comparison(df, cat_col: str, delay_col: str = "Dep_Delay_Tag"):
    """
    Vẽ biểu đồ so sánh giữa các nhóm Delay / No Delay theo biến phân loại cat_col.
    - df: DataFrame
    - cat_col: tên biến phân loại (ví dụ "Day_Of_Week", "Airline", ...)
    - delay_col: tên cột nhãn delay (mặc định = "Dep_Delay_Tag")
    """

    plt.figure(figsize=(8, 5))

    sns.countplot(
        data=df,
        x=cat_col,
        hue=delay_col,       # tạo 2 cột cạnh nhau (0, 1)
        palette="Set2"
    )

    plt.title(f"Delay vs Non-Delay by {cat_col}", fontsize=13, fontweight="bold")
    plt.xlabel(cat_col)
    plt.ylabel("Số chuyến bay")
    plt.legend(title="Delay Tag", labels=["0 = No Delay", "1 = Delay"], loc="upper right")

    plt.tight_layout()
    plt.show()
    
plot_delay_comparison(df, "Day_Of_Week")
plot_delay_comparison(df, "Manufacturer")
plot_delay_comparison(df, "month")
plot_delay_comparison(df, "Distance_type")
plot_delay_comparison(df, "DepTime_label")
```

```python
df_daily = df.groupby("FlightDate")["Dep_Delay"].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(x="FlightDate", y="Dep_Delay", data=df_daily, color="#A868D2")
plt.title("Average Departure Delay Over Time")
plt.xlabel("Date")
plt.ylabel("Average Delay (minutes)")
plt.grid ()
plt.tight_layout()
plt.show()

df_daily_2 = df.groupby("FlightDate")["Arr_Delay"].mean().reset_index()

plt.figure(figsize=(12,5))
sns.lineplot(x="FlightDate", y="Arr_Delay", data=df_daily_2, color="#63BDAD")
plt.title("Average Arrival Delay Over Time")
plt.xlabel("Date")
plt.ylabel("Average Delay (minutes)")
plt.grid ()
plt.tight_layout()
plt.show()
```

```python
df_daily["rolling_delay"] = df_daily["Dep_Delay"].rolling(window=7).mean()

plt.figure(figsize=(12,5))
sns.lineplot(x="FlightDate", y="rolling_delay", data=df_daily, color="#FF8C00")
plt.title("7-Day Rolling Average of Departure Delay")
plt.xlabel("Date")
plt.ylabel("Rolling Average Delay (minutes)")
plt.grid ()
plt.tight_layout()
plt.show()

df_daily_2["rolling_delay"] = df_daily_2["Arr_Delay"].rolling(window=7).mean()

plt.figure(figsize=(12,5))
sns.lineplot(x="FlightDate", y="rolling_delay", data=df_daily_2, color="#F04D7E")
plt.title("7-Day Rolling Average of Arrival Delay")
plt.xlabel("Date")
plt.ylabel("Rolling Average Delay (minutes)")
plt.grid ()
plt.tight_layout()
plt.show()
```
